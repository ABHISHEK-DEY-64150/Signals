***Compute linear convolution of two signals. Compare the result with the output of the
Matlab/Python library function.***

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

h = [1,2,3,1,1,1]
x = [1,2,3,4,5]
N1 = len(x)
N2 = len(h)
N = N1+N2-1
y = np.zeros(N)
p = np.arange(0,N)
m = N-N1
n = N-N2
x =np.pad(x,(0,m),'constant')
h =np.pad(h,(0,n),'constant')

for n in range (N):
    for k in range (N):
        if n >= k:
             y[n] = y[n]+x[n-k]*h[k]
         

print('Linear convolution using convolution  output response y =\n',y)


plt.figure(figsize=(20, 5))

plt.stem(p,h)

plt.title(' Impulse Response ->h', fontsize = 15)

plt.xlabel('Time')

plt.ylabel('Amplitude')

plt.show()


plt.figure(figsize=(20, 5))

plt.stem(p,x)

plt.title(' Input Signal ->x', fontsize = 15)

plt.xlabel('Time')

plt.ylabel('Amplitude')

plt.show()





plt.figure(figsize=(20, 5))

plt.stem(p,y)

plt.title(' Convolved Signal', fontsize = 15)

plt.xlabel('Time')

plt.ylabel('Amplitude')

plt.show()



```
