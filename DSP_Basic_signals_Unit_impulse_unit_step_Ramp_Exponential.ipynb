**# Unit impulse**

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = np.arange(-20,20)

l = np.size(n)

imp = np.zeros(l)

index = np.where(n==0)

imp[index] = 1


plt.figure(figsize=(20, 5))

plt.plot(n,imp)

plt.title(' Continuous Impulse Signal', fontsize = 15)

plt.xlabel('Time')

plt.ylabel('Amplitude')

plt.show()


plt.figure(figsize=(20, 5))

plt.stem(n,imp)

plt.title('Discrete Impulse Signal', fontsize = 15)

plt.xlabel('Sample Number')

plt.ylabel('Amplitude')

plt.show()







```

**#Ramp Signal**

```python
n = np.arange(-20,20)

rmp = []

tmp = 0

for i in range(len(n)):
    temp = (n[i] if n[i]>=0 else 0)
    rmp.append(temp)
    
plt.figure(figsize=(20, 5))

plt.plot(n,rmp)

plt.title('Continuous Unit Ramp Signal', fontsize = 15)

plt.xlabel('Time')

plt.ylabel('Amplitude')

plt.show()


plt.figure(figsize=(20, 5))

plt.stem(n,rmp)

plt.title('Discrete Unit Ramp Signal', fontsize = 15)

plt.xlabel('Sample number')

plt.ylabel('Amplitude')

plt.show()
```

**#Exponential Signal**

```python
a = 1.1
n = np.arange(-20,20)
exp = []

for i in range(len(n)):
    temp = a**n[i]
    exp.append(temp)
   

plt.figure(figsize=(20, 5))

plt.plot(n,exp)

plt.title('Continuous Exponential Signal a>0', fontsize = 15)

plt.xlabel('Time')

plt.ylabel('Amplitude')

plt.show()  


plt.figure(figsize=(20, 5))

plt.stem(n,exp)

plt.title('Discrete Exponential Signal a>0', fontsize = 15)

plt.xlabel('Sample number')

plt.ylabel('Amplitude')

plt.show()



a = -1.1
n = np.arange(-20,20)
exp = []

for i in range(len(n)):
    temp = a**n[i]
    exp.append(temp) 


plt.figure(figsize=(20, 5))

plt.plot(n,exp)

plt.title('Continuous Exponential Signal a<0', fontsize = 15)

plt.xlabel('Time')

plt.ylabel('Amplitude')

plt.show()  


plt.figure(figsize=(20, 5))

plt.stem(n,exp)

plt.title('Discrete Exponential Signal a<0', fontsize = 15)

plt.xlabel('Sample number')

plt.ylabel('Amplitude')

plt.show()    
```
